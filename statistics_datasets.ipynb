```python
import pandas as pd
import matplotlib.pyplot as plt
```

# Calculate stats for the full TA-ORF dataset (after QC)

## Calculate wells/sites/cells per pathway

```python
taorf_meta = pd.read_csv("/raid/data/cellpainting/TAORF/inputs/metadata/index_new_taorf_minus6wells.csv")
cells_taorf = []
for i in taorf_meta.index:
    cells_taorf.append([
        taorf_meta.loc[i, "Metadata_Plate"],
        taorf_meta.loc[i, "Metadata_Well"],
        taorf_meta.loc[i, "Metadata_Site"],
        sum(1 for line in open('/raid/data/cellpainting/TAORF/inputs/locations/{}/{}-{}-Nuclei.csv'.format(taorf_meta.loc[i, "Metadata_Plate"], taorf_meta.loc[i, "Metadata_Well"], taorf_meta.loc[i, "Metadata_Site"]))) - 1
    ])
counts = pd.DataFrame(data=cells_taorf, columns=['Metadata_Plate', 'Metadata_Well', 'Metadata_Site', 'Cells'])
taorf_meta = pd.merge(taorf_meta, counts, right_on=['Metadata_Plate', 'Metadata_Well', 'Metadata_Site'], left_on =['Metadata_Plate', 'Metadata_Well', 'Metadata_Site'])

taorf_meta
```

```python
taorf_moa_matches = pd.read_csv("/raid/data/cellpainting/TAORF/nikita_experiments/TAORF_MOA_MATCHES.csv")
taorf_meta_moa = pd.merge(taorf_meta, taorf_moa_matches[['Var1', 'Metadata_moa.x']], right_on=['Var1'], left_on =['broad_sample'])
taorf_meta_moa
```

## Calculate perturbations per pathway

```python
pert_taorf_counts = taorf_meta_moa[['Metadata_moa.x','pert_name']].drop_duplicates().groupby(['Metadata_moa.x']).size()
plt.hist(pert_taorf_counts, bins = 50)
```

## Calculate single cells per well/site/perturbation

```python
plt.figure()
plt.hist(taorf_meta.groupby(['Metadata_Plate','Metadata_Well']).agg('sum')['Cells'], bins = 50)
plt.figure()
#Cut EMPTY, 
plt.hist(taorf_meta[~taorf_meta['pert_name'].isin(['EMPTY_', 'Luciferase_CTRL', 'LacZ_CTRL', 'eGFP_CTRL'])].groupby(['pert_name']).agg('sum')['Cells'], bins = 50)
plt.figure()
plt.hist(taorf_meta_moa.groupby(['Metadata_moa.x']).agg('sum')['Cells'], bins = 50)
```

# Calculate stats for the full BBBC022 dataset (after QC)

```python
bbbc022_meta = pd.read_csv("/dgx1nas1/cellpainting-datasets/BBBC022/inputs/metadata/index_after_qc.csv")
#the number of cells is inferred from locations
cells_bbbc022 = []
for i in bbbc022_meta.index:
    cells_bbbc022.append([
        bbbc022_meta.loc[i, "Metadata_Plate"],
        bbbc022_meta.loc[i, "Metadata_Well"],
        bbbc022_meta.loc[i, "Metadata_Site"],
        sum(1 for line in open('/dgx1nas1/cellpainting-datasets/BBBC022/inputs/locations/{}/{}-{}-Nuclei.csv'.format(bbbc022_meta.loc[i, "Metadata_Plate"], bbbc022_meta.loc[i, "Metadata_Well"], bbbc022_meta.loc[i, "Metadata_Site"]))) - 1
    ])
counts = pd.DataFrame(data=cells_bbbc022, columns=['Metadata_Plate', 'Metadata_Well', 'Metadata_Site', 'Cells'])
bbbc022_meta = pd.merge(bbbc022_meta, counts, right_on=['Metadata_Plate', 'Metadata_Well', 'Metadata_Site'], left_on =['Metadata_Plate', 'Metadata_Well', 'Metadata_Site'])
```

```python
bbbc022_moa_matches = pd.read_csv("./bbbc022/data/BBBC022_MOA_MATCHES_fixed.csv")
bbbc022_moa_matches = bbbc022_moa_matches.assign(var1=bbbc022_moa_matches['Metadata_moa.x'].str.split('|')).explode('var1')
bbbc022_moa_matches.reset_index(inplace=True, drop = True)
bbbc022_moa_matches.drop(columns=['Unnamed: 0', 'Unnamed: 0.1', 'Metadata_moa.x'], inplace= True)
bbbc022_moa_matches.rename(columns={"var1": "Metadata_moa.x"}, inplace=True)

bbbc022_meta_moa = pd.merge(bbbc022_meta, bbbc022_moa_matches[['Var1', 'Metadata_moa.x']], right_on=['Var1'], left_on =['Compound'])
bbbc022_meta_moa
```

## Calculate wells/sites/cells per MoA

```python
site_moa_counts = bbbc022_meta_moa[['Metadata_moa.x','Metadata_Plate', 'Metadata_Well', 'Metadata_Site']].drop_duplicates().groupby(['Metadata_moa.x']).size()
plt.figure()
plt.hist(site_moa_counts, bins = 50)
well_moa_counts = bbbc022_meta_moa[['Metadata_moa.x','Metadata_Plate', 'Metadata_Well']].drop_duplicates().groupby(['Metadata_moa.x']).size()
plt.figure()
plt.hist(well_moa_counts, bins = 50)
```

## Calculate treatments per MoA

```python
compound_bbbc022_counts = bbbc022_meta_moa[['Metadata_moa.x','Treatment']].drop_duplicates().groupby(['Metadata_moa.x']).size()
#compound_bbbc022_counts[compound_bbbc022_counts>30]
plt.hist(compound_bbbc022_counts, bins = 50)
```

## Calculate single cells per well/treatment/MoA

```python
plt.figure()
plt.hist(bbbc022_meta.groupby(['Metadata_Plate','Metadata_Well']).agg('sum')['Cells'], bins = 50)
plt.figure()
#Cut DMSO as it has 1144258 cells 
plt.hist(bbbc022_meta[~bbbc022_meta['Treatment'].isin(['DMSO@0'])].groupby(['Treatment']).agg('sum')['Cells'], bins = 50)
plt.figure()
plt.hist(bbbc022_meta_moa.groupby(['Metadata_moa.x']).agg('sum')['Cells'], bins = 50)
```
